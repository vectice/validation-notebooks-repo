#### Retrieve Datasets

```python
import pandas as pd

# Loads the training and testing datasets from the specified paths in the configuration file
train_df = pd.read_csv(config_file["validation_config"]["paths"]["train"])
test_df = pd.read_csv(config_file["validation_config"]["paths"]["test"])
```

### Connect to Vectice to retrieve Model Version and download attachements

```python
import vectice

connection = vectice.connect(api_token=config_file["api_token"],host=config_file["host"],)

# Retrieve the Model version to be tested from Vectice to append to download attachments and pickle
# Retrieve the model version to be tested
model_version =  connection.browse(config_file["validation_config"]["model"])

# Define the directory where attachments should be downloaded
output_directiory_for_attachement = "./"

# Download attachments
model_version.download_attachments(output_dir=output_directiory_for_attachement)
```

#### Retrieve Model Pickle

```python
import pickle
# Loads the pre-trained model from the specified path in the configuration file
with open(output_directiory_for_attachement+config_file["validation_config"]["predictor"], "rb") as file:
    predictor = pickle.load(file)
```

### 3.1 Log the model you are testing to Vectice


```python
from vectice import autolog

# Configures the connect to Vectice
iteration = connection.browse(config_file['phase']).create_iteration()

## The model version currently tested
iteration.log(model_version, section = "Model tested")
```

### 3.2 Use modeva wrappers to create a factsheet that can trigger tests

```python
from vectice.models.wrapper.modeva_wrapper import modeva_wrapper

factsheet = modeva_wrapper(train_df, test_df, main_model=predictor, target=config_file["validation_config"]["target"])
```

### 3.3 Run you tests
#### 3.3.1 EDA

```python
result_eda = factsheet.get_dataset().eda_1d(feature="TARGET", plot_type="histogram")
# Display chart on screen
result_eda.plot(figsize=(6,4))
iteration.log(result_eda, section="eda_1d")
```

#### 3.3.2 Explainability

```python
result_shap = factsheet.explain_shap()
result_shap.plot(figsize=(6, 4))
iteration.log(result_shap, section="explain_shap")
```

#### 3.3.3 Performance

```python
result_diag = factsheet.diagnose_accuracy_table()
result_diag.plot(figsize=(6, 4))
iteration.log(result_diag)
```

#### 3.3.4 Robustness, Resilience and reliability

```python
result_robustness = factsheet.diagnose_robustness()
result_robustness.plot(figsize=(6, 4))
iteration.log(result_robustness)
```

```python
result_reliability = factsheet.diagnose_reliability()
result_reliability.plot(figsize=(6, 4))
iteration.log(result_reliability)
```

```python
result_resilience = factsheet.diagnose_resilience()
result_resilience.plot(figsize=(6, 4))
iteration.log(result_resilience)
```
