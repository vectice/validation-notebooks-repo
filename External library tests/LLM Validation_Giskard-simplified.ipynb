## 2 - Initialize Scan Framework

In order to run the scanner, the framework requires the following to be initialized:
 - a chat engine leveraging the LLM client
 - the set of vulnerabilities to scan against


### Initialization of a Retrieval-Augmented Generation (RAG) Framework

This code block sets up a RAG framework by performing the following steps:

1. **Embedding Model Setup**: Configures the global embedding model using Azure OpenAI for efficient text embedding.
2. **Index Rehydration**: Unpacks and loads a previously stored index, which facilitates quick retrieval of relevant information from storage.
3. **LLM Client Initialization**: Initializes the Azure OpenAI client using API credentials, preparing the environment for interaction with the language model.
4. **LLM Configuration**: Sets up a specific language model (GPT-3.5-turbo), with parameters that control its behavior, such as deterministic output generation.
5. **Chat Engine Creation**: Combines the loaded index and the configured LLM to create a chat engine, enabling natural language queries that leverage both retrieval (from the index) and generation (from the LLM).

This setup forms the foundation of a RAG system, where the index is used to retrieve relevant context, and the LLM generates responses based on this context.

---

**Note for Model Validators:**

The steps above primarily concern the setup and configuration of the RAG framework, typically handled by the modeling team. To streamline the process, we've encapsulated these steps into a function that initializes the LLM. **Model validators do not need to focus on this setup process.** Their task will mainly involve validating the outputs generated by the pre-configured LLM within the RAG framework, rather than engaging with the underlying setup code.

---

## 3 - Scan with Giskard (Run your tests) -- Validator part

#### Focus on LLM Scanning

In this section, we focus on the scanning capabilities of the LLM within Giskard, which helps identify vulnerabilities and ensure model safety.

**Beyond Scanning:** Giskard also offers advanced testing strategies, including adversarial testing and bias detection, as well as RAG (Retrieval-Augmented Generation) reports for a more comprehensive model evaluation.

#### Load the testing set

```python
import giskard
from giskard.rag import QATestset

scanset = QATestset.load("./ipcc_scanset.jsonl")
giskard_dataset = giskard.Dataset(df=scanset.to_pandas()[["question"]])
```

#### Create wrapper Giskard Model

```python
import giskard

giskard_model = giskard.Model(
    model=model_predict,
    model_type="text_generation",
    name="Climate Change Question Answering",
    description="This model answers any question about climate change based on IPCC reports",
    feature_names=["question"],  # input variables needed by your model
)
```

#### Configure Vulnerabilities to Scan Against

The list of vulnerabilities can be configured through the `config.json` file or a git-versioned file. Note that the current setup includes a short list of vulnerabilities to optimize test runtime.

**Note for Validators:** You have the flexibility to adjust the type of scan you wish to test by modifying the configuration. Depending on your validation needs, you can expand or refine the list of vulnerabilities for more comprehensive or targeted testing.

```python
vulnerabilities = ["information_disclosure", "hallucination"]# 2 test types --> sections 
```

#### Execute the Scan

```python
# Occasionally, the information_disclosure process may fail.
# This usually happens due to issues with the API call to ChatGPT.
# If you encounter this failure, try rerunning the cell.
scan_results = giskard.scan(giskard_model,giskard_dataset ,only=vulnerabilities)
```
