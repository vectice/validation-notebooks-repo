### 3.1 Roc curve reproducibility

```python
## Test 1: Roc curve plot reproducibility
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import auc, roc_curve


def plot_roc_curve(
    training_df,
    testing_df,
    target_column,
    predictor,
    internal_parameters = {"train_color": "green", "test_color": "blue", "threshold": 0.5},
) -> None:
    X_train = training_df.drop(columns=[target_column])
    X_test = testing_df.drop(columns=[target_column])
    training_prediction_proba = predictor.predict_proba(X_train)[:, 1]
    testing_prediction_proba = predictor.predict_proba(X_test)[:, 1]

    fpr_train, tpr_train, _ = roc_curve(training_df[target_column], training_prediction_proba)
    roc_auc_train = auc(fpr_train, tpr_train)

    fpr_test, tpr_test, _ = roc_curve(testing_df[target_column], testing_prediction_proba)
    roc_auc_test = auc(fpr_test, tpr_test)

    file_path = "Reproduced_ROC_CURVE.png"

    plt.figure(figsize=(8, 6))
    plt.plot(
        fpr_train,
        tpr_train,
        color=internal_parameters["train_color"],
        linestyle="--",
        label=f"Train ROC curve (AUC = {roc_auc_train:.2f})",
    )
    plt.plot(
        fpr_test,
        tpr_test,
        color=internal_parameters["test_color"],
        label=f"Test ROC curve (AUC = {roc_auc_test:.2f})",
    )
    plt.plot([0, 1], [0, 1], color="red", linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver Operating Characteristic (ROC) Curve")
    plt.legend()
    plt.grid(True)
    plt.savefig(file_path)
    plt.show()
    plt.close()
```

```python
X_test = test_df.drop(columns=["TARGET"]).values
y_test = test_df["TARGET"]

test_predictions = predictor.predict_proba(X_test)[:, 1]
```

```python
plot_roc_curve(training_df=train_df, testing_df=test_df, target_column="TARGET", predictor=predictor)
```

### 3.2 Metrics by threshold

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def generate_table_by_threshold(y_true, y_pred_prob):
    results = pd.DataFrame(columns=['Threshold', 'Accuracy', 'Precision', 'Recall'])
    
    thresholds = np.arange(0.0, 1.0, 0.1)
    thresholds = np.round(thresholds, 1)
    
    
    for t in thresholds:
        y_pred = (y_pred_prob >= t).astype(int)
        accuracy = accuracy_score(y_true, y_pred)
        
        # Handle precision and recall to avoid undefined metrics
        if y_pred.sum() == 0:
            precision = 0.0
        else:
            precision = precision_score(y_true, y_pred)
        
        if y_true.sum() == 0:
            recall = 0.0
        else:
            recall = recall_score(y_true, y_pred)
        
        new_results = pd.DataFrame({
            'Threshold': [t],
            'Accuracy': [accuracy],
            'Precision': [precision],
            'Recall': [recall]
        })
        results = pd.concat([results, new_results], ignore_index=True)

    return results
```

```python
from vectice import Table
## Test 2: Metrics by threshold
results = generate_table_by_threshold(y_test.values, test_predictions)
## This model is not calibrated properly
threshold_table = Table(results, name="Metrics by thresholds")
threshold_table.dataframe
```
